```python
bs_url = "https://store.steampowered.com/games/#p=0&tab=TopSellers"
html=urlopen(bs_url)
soup=BeautifulSoup(html, "lxml")
```

```python
#게임이 몇 개인지 알아낸다.
number=soup.find("span",{"id":"TopSellers_total"}) 
number=number.contents[0]
number=number.replace(",","")
number=int(number)
print(number, "개의 게임이 조회되었습니다. \n")
```

```python
title_list=[]
```

```python
for i in list(range(0,3)):
    page_url="https://store.steampowered.com/games/#p={}&tab=TopSellers".format(i)
    driver.get(page_url)
    driver.refresh()
    time.sleep(2)
    soup=BeautifulSoup(driver.page_source, "lxml")
    all_source=soup.find("div",{"id":"tab_content_TopSellers"})
    title=all_source.find_all("div",{"class":"tab_item_name"})
    title_list.append(title)
    print(i)
```

```python
title_list
```

```python
title_list=sum(title_list, [])
```

```python
title_list
```

```python
len(title_list)
```

```python
best_selling=[]
```

```python
for i in title_list:
    best_selling.append(i.string)
```

```python
best_selling
```

```python
import pandas as pd
```

```python
dataframe=pd.DataFrame(best_selling)
dataframe
```
